```python
"""
수집된 데이터 중에 중복이 있는 경우, 해당 중복 건에 대하여 열 값들을 합치고, 데이터 출처를 이어서 적도록 하는 함수 
"""

def duplicate_aggregate(df):
    
    dup_idx = []
    for i in df.index:
        if not list(df.index).count(i) == 1:
            dup_idx.append(i)
#     assert dup_idx != set()  
    if len(dup_idx) != 0:

        dup_idx = set(dup_idx)
        new_df = pd.DataFrame(columns = df.columns)

        for i, dup in zip(range(len(dup_idx)), dup_idx):
            attr = df.loc[dup, df.columns[:-1]].values
            attr = attr[0] + attr[1]
            attr = np.clip(attr, a_min = 0, a_max = 1)
            sources = list(set(df.loc[dup, 'source'].values))
            if len(sources) > 1:
                sources = ', '.join(sources)
            else:
                sources = sources[0]
            new_df.loc[dup] = list(attr) + [sources]
            df = df.drop(dup)
        return df.append(new_df, sort = False) 
    
    elif len(dup_idx) == 0:
        return df
```

```python
df2 = df1.append(df, sort = False)
df2 = duplicate_aggregate(df2)
```
