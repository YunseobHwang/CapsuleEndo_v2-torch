## 3.2. Accr by sub-class

```python
def test_by_subclass(model, filename, X, Y):
    def extract_accr(model, x, y):
#     accrs = []
#     for i, test_x in enumerate(test_xs):
#         if i == 0:
#             model.Evaluation(test_x, GenerateLabel(test_x, 0))
#         else:
#             model.Evaluation(test_x, GenerateLabel(test_x, 1))
#         accrs.append(model.accr)
    
        model.evaluation(x, y, batch_size = 20, print_accr = False)
        return model.accr
    
    label_dir = '/mnt/disk2/data/private_data/SMhospital/capsule/1 preprocessed'
    label = pd.read_csv(label_dir + '/label.csv', index_col = 0)
    
    accrs = []
    for sub_cls in label.columns[:-1]:
    
        x, y = [], []
        for name, img, cls in zip(filename, X, Y):
            if label.loc[('').join(name.split('__c_-_-_-'))][sub_cls] == 1:
                x.append(img)
                y.append(cls)

        x = np.concatenate([x])
        y = np.asarray(y)
    
        accrs.append(extract_accr(model, x, y)) 
    return accrs
```

# 4. Discussion

- H와 D로만 구성된 데이터(np_hd)와 protruded, etc까지 전체 lesion(abnormal)까지 포함한 데이터(np)에 대해
- 각각 훈련한 모델로 각각의 데이터 테스트셋으로 평가한 결과,
- 이전에 H, D로만 훈련한 r, f + r + b가 가장 좋은 성능 보였다.
- 데이터 N수가 증가해 모든 데이터 증강 방법을 사용하기엔 훈련시간이 너무 길어 r만 선택했다
----------------------------------------------------------------------------------------------
- feedback 데이터는 라벨링이 그냥 positive로 hemorrhagic (H), depressed (D) 만 고려된 것이 아님
- H가 D보다 쉬웠던 문제로 알고 있는데, negative로부터 구분하기 어려운 데이터가 들어간 것으로 생각되어진다. 
- AI 모델 입장에서 문제 난이도가 상승, H, D로만 훈련시켰던 모델보다 오히려 성능이 낮게 나온.. 멍청한 모델로 훈련된..느낌
- 개와 고양이를 구분하는데 개, 고양이 중간 같이 생긴게 들어가서
- 개, 고양이를 더 구분 못 하게되는 원리
------------------------------------------------------------------------------------------------

- 그렇다면 모든 종류의 병변을 positive로 두고 훈련하는 방향 실제적으로 적용하기 위한 방향으로 가야한다

- 모르는 종류(sub-class - abnormal category) open class의 abnormal데이터가 모델 훈련에 들어갔을 수도 있다
