```python
sources = sorted(list(set(label.source.values)))
target_sources = [source for source in sources if '200713 clip for AI re-screening' in source]

for t_s in target_sources:
    rename = t_s.replace('200713 clip for AI re-screening', '200713-1 AI feedback')
    label = label.replace({'source': {t_s: rename}})
```
