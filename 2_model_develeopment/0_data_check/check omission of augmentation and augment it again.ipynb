```python
import time
from tqdm import tqdm
```

```python
from ce_utils.record import progress_bar, sec_to_m_s_ms
```

```python
data_dir = '/mnt/disk2/data/private_data/SMhospital/capsule/1 preprocessed/database'

omission = ['byon jeong suk (29675085) 06 Jul 17_1_frame_12551__c_-_-_-.jpg', 
            'byon jeong suk (29675085) 06 Jul 17_1_frame_15077__c_-_-_-.jpg']

start = time.time()
pbar = tqdm(total=len(omission), unit='id', bar_format='{l_bar}{bar:40}{r_bar}')

for i, file in enumerate(omission):
    
    pbar.update(1)
    
    img = cv2.imread(os.path.join(data_dir, file))
    aug_suffixes = extract_aug_suffix(frb_switch = [1, 1, 1], sv_switch = True, mode = 'load')
    for aug_suffix in aug_suffixes:
        aug_img = pre_aug(img, aug_suffix)
        cv2.imwrite(data_dir + '/' + file.split('-_-_-')[0] + aug_suffix + '.jpg', aug_img)
                    
taken = time.time()-start
print()
print(sec_to_m_s_ms(taken), end ="\n")
```
