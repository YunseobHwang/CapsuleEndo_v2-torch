```python
import pandas as pd
import os, pickle
from glob import glob
import numpy as np

root = '/mnt/disk2/data/private_data/SMhospital/capsule/1 preprocessed'

label = pd.read_csv(root + '/label.csv', index_col = 0)
label
```

```python
paths = []
for raw_file in label.index.values:
    if not 'annotation' in raw_file:
         paths.append(root + '/database/' + raw_file.split('.jpg')[0] + '__c_-_-_-.jpg')
    elif 'annotation' in raw_file:
        paths.append(root + '/database/' + raw_file.split('_annotation')[0] + '__c_-_-_-.jpg')
```
