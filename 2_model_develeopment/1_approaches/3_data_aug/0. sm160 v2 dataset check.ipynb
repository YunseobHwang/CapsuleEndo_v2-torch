```python
from CE_utils import sm160_dataset
import os
import numpy as np

sm160_train = sm160_dataset(phase = 'train', data = 'sm_x160_v2', pre_a = [0, 0, 0], pre_b = False, ext_name = False)
sm160_test = sm160_dataset(phase = 'test', data = 'sm_x160_v2', pre_a = [0, 0, 0], pre_b = False, ext_name = False)

neg_train_path = sm160_train.load_path('n', 'neg')
redspot_train_path = sm160_train.load_path('h', 'redspot')
angio_train_path = sm160_train.load_path('h', 'angio')
active_train_path = sm160_train.load_path('h', 'active')
ero_train_path = sm160_train.load_path('d', 'ero')
ulcer_train_path = sm160_train.load_path('d', 'ulcer')
str_train_path = sm160_train.load_path('d', 'str')

neg_test_path = sm160_test.load_path('n', 'neg')
redspot_test_path = sm160_test.load_path('h', 'redspot')
angio_test_path = sm160_test.load_path('h', 'angio')
active_test_path = sm160_test.load_path('h', 'active')
ero_test_path = sm160_test.load_path('d', 'ero')
ulcer_test_path = sm160_test.load_path('d', 'ulcer')
str_test_path = sm160_test.load_path('d', 'str')

neg_train_path = [os.path.basename(i) for i in neg_train_path]
redspot_train_path = [os.path.basename(i) for i in redspot_train_path]
angio_train_path = [os.path.basename(i) for i in angio_train_path]
active_train_path = [os.path.basename(i) for i in active_train_path]
ero_train_path = [os.path.basename(i) for i in ero_train_path]
ulcer_train_path = [os.path.basename(i) for i in ulcer_train_path]
str_train_path = [os.path.basename(i) for i in str_train_path]

neg_test_path = [os.path.basename(i) for i in neg_test_path]
redspot_test_path = [os.path.basename(i) for i in redspot_test_path]
angio_test_path = [os.path.basename(i) for i in angio_test_path]
active_test_path = [os.path.basename(i) for i in active_test_path]
ero_test_path = [os.path.basename(i) for i in ero_test_path]
ulcer_test_path = [os.path.basename(i) for i in ulcer_test_path]
str_test_path = [os.path.basename(i) for i in str_test_path]

neg_duplicate = np.intersect1d(neg_train_path, neg_test_path)
redspot_duplicate = np.intersect1d(redspot_train_path, redspot_test_path)
angio_duplicate = np.intersect1d(angio_train_path, angio_test_path)
active_duplicate = np.intersect1d(active_train_path, active_test_path)
ero_duplicate = np.intersect1d(ero_train_path, ero_test_path)
ulcer_duplicate = np.intersect1d(ulcer_train_path, ulcer_test_path)
str_duplicate = np.intersect1d(str_train_path, str_test_path)

print(neg_duplicate)
print(redspot_duplicate)
print(angio_duplicate)
print(active_duplicate)
print(ero_duplicate)
print(ulcer_duplicate)
print(str_duplicate)
```
