## 2.2. Accuracy, Sensitivity ...

```python
from sklearn.metrics import confusion_matrix

class cls_metric:
    def __init__(self, data_y, score):
        self.true = data_y 
        self.pred = np.argmax(score, axis = 1)
        self.accr = 100*np.mean(np.equal(self.true, self.pred))
        self.cm = confusion_matrix(self.true, self.pred)

    def plot_cm(self, value_size = 15, label_size = 10, mode = 'percent'):
        plt.imshow(self.cm, interpolation = 'nearest', cmap = plt.cm.Blues)
        thresh = self.cm.max()/2.
        for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            if mode == 'percent':
                value = np.round(cm[i, j]/(np.sum(cm, 1)[i]), 3)
            if mode == 'num':
                value = cm[i, j]
            plt.text(j, i, value,
                     fontsize = value_size,
                     horizontalalignment = 'center',
                     color = 'white' if cm[i, j] > thresh else 'black')
        plt.ylabel('True label', fontsize = label_size)
        plt.xlabel('Predicted', fontsize = label_size)
        plt.xticks([0, 1], ['nor', 'abnor'], rotation=0, fontsize = label_size)
        plt.yticks([0, 1], ['nor', 'abnor'], rotation=90, fontsize = label_size)

    def cm2metric(self):
        """
        A confusion matrix for Binary classification
        2x2 array: [[TN, FN],[FP, TP]]
        """
        Acc = round(100*(self.cm[0, 0] + self.cm[1, 1]) / np.sum(self.cm), 2)
        Sen = round(100*self.cm[1, 1] / (self.cm[1, 0] + self.cm[1, 1]), 2)
        Spec = round(100*self.cm[0, 0] / (self.cm[0, 0] + self.cm[0, 1]), 2)
        NPV = round(100*self.cm[0, 0] / (self.cm[0, 0] + self.cm[1, 0]), 2)
        PPV = round(100*self.cm[1, 1] / (self.cm[0, 1] + self.cm[1, 1]), 2)
        return [Acc, Sen, Spec, NPV, PPV]

    def confusion_idx(self, cls = 'positive'):
        if cls == 'positive':
            fn_idx = np.where(self.true - self.pred == 1)[0]
            tp_idx = np.setxor1d(np.where(self.true == 1), fn_idx)
            return tp_idx, fn_idx
        elif cls == 'negative':
            fp_idx = np.where(self.true - self.pred == -1)[0]
            tn_idx = np.setxor1d(np.where(self.true == 0), fp_idx)
            return tn_idx, fp_idx
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
def ROCplot(labels, outputs, 
            xlim, ylim,  
            label_name = '', axis_line = None):
    fpr, tpr, _ = roc_curve(labels, outputs[:, 1])
    roc_auc = auc(fpr, tpr)
        
    gmeans = np.sqrt(tpr * (1-fpr))
    idx = np.argmax(gmeans)
    
    height = ylim[1] - ylim[0]
    width = xlim[1] - xlim[0]
    
    ax.plot(fpr, tpr, lw = 2, label='{} (AUC = {:.4f})'.format(label_name, roc_auc))
#     ax.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
#     ax.plot(fpr[idx], tpr[idx], 'o', mfc = color, mec = color, ms = 3)
#     ax.plot(fpr[idx], tpr[idx], 'o', ms = 3)
    if axis_line is not None:
        ax.axvline(fpr[idx], color = 'k', linestyle='dashed', linewidth=1, ymax=(tpr[idx]- ylim[0])/height)
        ax.axhline(tpr[idx], color = 'k', linestyle='dashed', linewidth=1, xmax=(fpr[idx]- xlim[0])/width )
```

- 모델들을 비교 평가한 결과, data augmenation은 모델의 성능을 향상시켰다. 
- (f--) flip한 건 오히려 generalization 성능을 떨어뜨렸고,
- r, f+r+b, f+r+b+sv가 가장 좋게 나타났다. 
- f+r+b+sv의 경우, auc가 왜 작은지 확인하기 위해 confidence를 살펴보았는데, 
- 그 부분이 굉장히 크다보니 오히려 auc가 줄어드는 현상이 나타났다.

- issue
    - 훈련시간이 뒤로 갈수록 데이터가 증강되니까 너무 오래걸림... 일주일내내 해도 안 끝남..
    - 매 epoch마다 원본 train dataset으로도 validation하는 걸 제거해야할듯 
    - 각 augmetaion방법마다 새로 불러와 train valid 나누게 되는데, 
    - valid_accr가 높은 걸 최우선 기준으로 하다보니 거기서 오는 bias가 있고 
    - 즉, valid dataset이 testing dataset과 비슷하게끔 구성되어있으면 그거에 fit한 모델이 test 성능이 높아짐.
    - train test이 random으로 나눠져있기는 했지만, bias가 존재하기 때문에 그 영향을 무시할 수 없고, 
    - strict하게 평가하기 위해선 train test를 random으로 나눠가며 multiple하게 평가하거나 k-fold cross validation
